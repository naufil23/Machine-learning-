```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = df.drop('deposit', axis=1)
y = df['deposit']
```

```python
categorical_cols = X.select_dtypes(include=['object']).columns
numerical_cols = X.select_dtypes(include=np.number).columns

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_encoded_categorical = pd.DataFrame(encoder.fit_transform(X[categorical_cols]))

X_encoded_categorical.columns = encoder.get_feature_names_out(categorical_cols)

X_encoded_categorical.index = X.index

X_processed = pd.concat([X[numerical_cols], X_encoded_categorical], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.3, random_state=42)

print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
print(f"Training target shape: {y_train.shape}")
print(f"Testing target shape: {y_test.shape}")
```

```python
cart_dt_model = DecisionTreeClassifier(random_state=42)

cart_dt_model.fit(X_train, y_train)

print("CART Decision Tree model trained successfully!")
```

```python
y_pred_cart = cart_dt_model.predict(X_test)

accuracy_cart = accuracy_score(y_test, y_pred_cart)
report_cart = classification_report(y_test, y_pred_cart)
conf_matrix_cart = confusion_matrix(y_test, y_pred_cart)

print(f"Accuracy of CART Decision Tree: {accuracy_cart:.4f}")
print("\nClassification Report (CART Decision Tree):\n")
print(report_cart)
print("\nConfusion Matrix (CART Decision Tree):\n")
print(conf_matrix_cart)
```

```python
id3_c45_dt_model = DecisionTreeClassifier(criterion='entropy', random_state=42)

id3_c45_dt_model.fit(X_train, y_train)

print("ID3/C4.5 Decision Tree model (Information Gain) trained successfully!")
```

```python
y_pred_id3_c45 = id3_c45_dt_model.predict(X_test)

accuracy_id3_c45 = accuracy_score(y_test, y_pred_id3_c45)
report_id3_c45 = classification_report(y_test, y_pred_id3_c45)
conf_matrix_id3_c45 = confusion_matrix(y_test, y_pred_id3_c45)

print(f"Accuracy of ID3/C4.5 Decision Tree: {accuracy_id3_c45:.4f}")
print("\nClassification Report (ID3/C4.5 Decision Tree):\n")
print(report_id3_c45)
print("\nConfusion Matrix (ID3/C4.5 Decision Tree):\n")
print(conf_matrix_id3_c45)
```

```python
print(f"CART (Gini Impurity) Model Accuracy: {accuracy_cart:.4f}")
print(f"ID3/C4.5 (Information Gain) Model Accuracy: {accuracy_id3_c45:.4f}")

print("\n--- CART (Gini) Classification Report ---")
print(report_cart)

print("\n--- ID3/C4.5 (Entropy) Classification Report ---")
print(report_id3_c45)
```
